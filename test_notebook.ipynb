```python
import numpy as np
```

```python
with open("data\\train-consonant-examples.txt", 'r') as text:
    data = text.readlines()
```

```python
print(data)
cleaned_data = []
for sentence in data:
    cleaned_data.append(sentence.strip('\n'))
```

```python
word_list = []
for i,sentence in enumerate(cleaned_data):
    words = sentence.split(" ")
    word_list.extend(words)
```

```python
len_list = []
for word in word_list:
    len_list.append(len(word))
```
